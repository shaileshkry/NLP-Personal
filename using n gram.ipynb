```python
def get_pairs(doc,n):
    data=[]
    for i in range(len(doc)-n):
        data.append(str(doc[i:i+1+n]))
    return data
```

```python
n=2
data=get_pairs(doc,n)
data
```

```python
unique_pairs=list(set(data))
```

```python
len(unique_pairs)
```

### Occurance probability

```python
prob_dist=[]
for pair in tqdm(unique_pairs):
    prob_dist.append([pair,data.count(pair)])
```
